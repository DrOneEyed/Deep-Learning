### Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import cross_val_score
```

### Reading Data

```python
df_housing = pd.read_csv("/content/sample_data/california_housing_train.csv")

X_train = df_housing[["housing_median_age", "total_rooms", "total_bedrooms"]]
y_train = df_housing["median_house_value"]
```

### Fitting Two Initial Models

```python
linear_model = LinearRegression()
linear_model.fit(X=X_train, y=y_train)

knn_model = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=10))
knn_model.fit(X=X_train, y=y_train)

new_house = [[1500, 2, 2]]
linear_model.predict(new_house), knn_model.predict(new_house)
```

### Voting Ensemble

```python
np.mean([linear_model.predict(new_house), knn_model.predict(new_house)])
ensemble_model = VotingRegressor([("linear", linear_model), ("knn", knn_model)])
ensemble_model.fit(X=X_train, y=y_train)
ensemble_model.predict(new_house)
```

### Cross-validation

```python
for model in [linear_model, knn_model, ensemble_model]:
    print(-cross_val_score(model, X=X_train, y=y_train, cv=5, scoring="neg_mean_squared_error").mean())
```
